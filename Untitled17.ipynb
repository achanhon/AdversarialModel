## objectif
L'objectif de ce notebook est d'illustrer la difficulté pour un réseau de neurones de produire une carte de segmentation sémantique avec des pics. En effet, les réseaux ont tendance à produire une sortie potentiellement trop lisse (donc des patates plutôt que des pics).

## methode
Pour cela, on peut considerer un problème jouet construit à partir d'un signal (ici les bords d'un carré de taille 7x7 associé avec une vérité terrain à 1 au niveau du pixel central).
On rajoute ensuite du bruit à l'image.

La carte qu'on veut obtenir est donc uniquement consistuée de 1 pixel blanc pour chaque carré et tout le reste est noir.
Cela est possible sans ambiguité car le niveau de bruit est inférieur au niveau du signal.
Pourtant, on va constater que les réseaux ont tendance à produire des patates et non des pics en fonction du niveau de bruit.

```python
import torch

def createSample():
  x = 0.9*torch.rand(3,128,128) # un bruit fort mais sans ambiguité 0.9<1
  y = torch.zeros(128,128)

  # add signal
  for k in range(3):
    tmp = torch.rand(2)*120
    j,i = int(tmp[0]),int(tmp[1])
    x[0,j:j+7,i]=1
    x[0,j:j+7,i+7]=1
    x[1,j,i:i+7]=1
    x[1,j+7,i:i+7]=1
    y[j+3][i+3]=1

  return x,y

import matplotlib.pyplot as plt

x,y = createSample()

y = torch.stack([y,y,y],dim=0)
xy = torch.cat([x,y],dim=-1)
xy = xy.cpu().numpy().transpose(1,2,0)

plt.imshow(xy)
plt.axis('off')
plt.show()
```

```python
class LittleUnet(torch.nn.Module):
  def __init__(self):
    super(LittleUnet, self).__init__()
    self.c11 = torch.nn.Conv2d(3, 32,kernel_size=3,padding=1)
    self.c12 = torch.nn.Conv2d(32, 32,kernel_size=3,padding=1)

    self.c21 = torch.nn.Conv2d(32, 64,kernel_size=3,padding=1)
    self.c22 = torch.nn.Conv2d(64, 64,kernel_size=3,padding=1)

    self.c31 = torch.nn.Conv2d(64, 128,kernel_size=3,padding=1)
    self.c32 = torch.nn.Conv2d(128, 128,kernel_size=3,padding=1)

    self.d11 = torch.nn.Conv2d(192, 64,kernel_size=3,padding=1)
    self.d12 = torch.nn.Conv2d(64, 64,kernel_size=3,padding=1)

    self.d21 = torch.nn.Conv2d(96, 32,kernel_size=3,padding=1)
    self.d22 = torch.nn.Conv2d(32, 32,kernel_size=3,padding=1)

    self.f = torch.nn.Conv2d(32, 2,kernel_size=1)

  def forward(self, x):
    x = torch.nn.functional.leaky_relu(self.c11(x))
    x = torch.nn.functional.leaky_relu(self.c12(x))

    x2 = torch.nn.functional.max_pool2d(x,kernel_size=2)
    x2 = torch.nn.functional.leaky_relu(self.c21(x2))
    x2 = torch.nn.functional.leaky_relu(self.c22(x2))

    x3 = torch.nn.functional.max_pool2d(x2,kernel_size=2)
    x3 = torch.nn.functional.leaky_relu(self.c31(x3))
    x3 = torch.nn.functional.leaky_relu(self.c32(x3))

    x3 = torch.nn.functional.interpolate(x3, size=(64,64), mode="bilinear")
    x2 = torch.cat([x2,x3],dim=1)
    x2 = torch.nn.functional.leaky_relu(self.d11(x2))
    x2 = torch.nn.functional.leaky_relu(self.d12(x2))

    x2 = torch.nn.functional.interpolate(x2, size=(128,128), mode="bilinear")
    x = torch.cat([x,x2],dim=1)
    x = torch.nn.functional.leaky_relu(self.d21(x))
    x = torch.nn.functional.leaky_relu(self.d22(x))

    return self.f(x)

net = LittleUnet().cuda()
criterion = torch.nn.CrossEntropyLoss(torch.Tensor([1,10]).cuda())
optimizer = torch.optim.Adam(net.parameters(), lr=0.00001)
printloss = 0
nbiter = 10000
```

```python
for i in range(nbiter):
  x, y = createSample()
  x, y = x.cuda().unsqueeze(0), y.cuda().unsqueeze(0)
  z = net(x)

  loss = criterion(z,y.long())

  with torch.no_grad():
    printloss += float(loss)

  optimizer.zero_grad()
  loss.backward()
  torch.nn.utils.clip_grad_norm_(net.parameters(), 1)
  optimizer.step()

  if i % 100==99:
    print(printloss/100)
    printloss = 0
```

```python
with torch.no_grad():
  cm = torch.zeros(2,2).cuda()
  for i in range(10):
    x, y = createSample()
    x, y = x.cuda().unsqueeze(0), y.cuda()
    z = net(x)[0]
    z = (z[1]>z[0]).float()
    for a,b in [(0,0),(0,1),(1,0),(1,1)]:
      cm[a][b] += ((z==a).float()*(y==b).float()).sum()
  # print("cm",cm)
  print("IoU",float(cm[1][1]/(cm[1][1]+cm[0][1]+cm[1][0]+1)))

  y = torch.stack([y,y,y],dim=0)
  z = torch.stack([z,z,z],dim=0)
  xyz = torch.cat([x[0],y,z],dim=-1)
  xyz = xyz.cpu().numpy().transpose(1,2,0)

  plt.imshow(xyz)
  plt.axis('off')
  plt.show()
```
